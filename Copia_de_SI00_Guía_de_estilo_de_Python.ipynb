```python
def contador_palabras(cadena_texto):
    contador=0
    palabras=cadena_texto.split()
    para palabra in palabras:
        contador+=1
    imprime("El número de palabras es:",contador)

cadena_texto=input("Ingrese una cadena de texto: ")
contador_palabras(cadena_texto)
```

```python
def contador_palabras(cadena_texto: str) -> int:
  """
  Esta función cuenta el número de palabras en una cadena de texto.

  Parámetros:
    cadena_texto (str): La cadena de texto a analizar.

  Retorno:
    int: El número de palabras en la cadena de texto.
  """
  # Se divide la cadena en palabras usando el separador por defecto (espacio en blanco).
  palabras = cadena_texto.split()

  # Se inicializa un contador de palabras.
  contador = 0

  # Se recorre la lista de palabras y se incrementa el contador en cada iteración.
  for _ in palabras:
    contador += 1

  return contador

# Se solicita al usuario que ingrese una cadena de texto.
cadena_texto = input("Ingrese una cadena de texto: ")

# Se llama a la función para contar las palabras en la cadena ingresada.
numero_palabras = contador_palabras(cadena_texto)

# Se imprime el resultado al usuario.
print(f"El número de palabras es: {numero_palabras}")
```

```python
def generar_primos(N):
    primos=[2]
    for num in range(3,N+1):
        for i in range(2,num):
            if(num%i)==0:
                break
        else:
            primos.append(num)
    print("Los primeros", N, "números primos son:")
    print(primos)

N=int(input("Ingrese el número de primos a generar: "))
generar_primos(N)
```

```python
def generar_primos(N: int) -> list:
  """
  Esta función genera una lista con los N primeros números primos.

  Parámetros:
    N (int): El número de primos a generar.

  Retorno:
    list: Una lista con los primeros N números primos.
  """

  # Se crea una lista para almacenar los números primos.
  primos = [2]

  # Se recorren los números desde 3 hasta N.
  for num in range(3, N + 1):
    # Se inicializa un booleano para indicar si el número es primo.
    es_primo = True

    # Se comprueba si el número es divisible por algún número entre 2 y num-1.
    for i in range(2, num):
      if num % i == 0:
        # Si el número es divisible por otro número, no es primo.
        es_primo = False
        # Se sale del bucle for interno.
        break

    # Si el número es primo, se añade a la lista.
    if es_primo:
      primos.append(num)

  # Se devuelve la lista de números primos.
  return primos

# Se solicita al usuario el número de primos a generar.
N = int(input("Ingrese el número de primos a generar: "))

# Se genera la lista de números primos.
primos = generar_primos(N)

# Se imprime la lista de números primos.
print(f"Los primeros {N} números primos son:")
print(primos)
```

```python
# Modifique en esta casilla el código anterior para que se ajuste al PEP8
"""
Módulo para generar contraseñas aleatorias.

Este módulo contiene una función para generar contraseñas aleatorias con una
longitud especificada por el usuario.

Funciones:
  generar_contraseña(longitud): Genera una contraseña aleatoria de la longitud
  especificada.
"""

import random
import string


def generar_contraseña(longitud):
  """
  Genera una contraseña aleatoria de la longitud especificada.

  Argumentos:
    longitud: La longitud de la contraseña a generar.

  Retorno:
    La contraseña generada.

  Excepciones:
    ValueError: Si la longitud de la contraseña es menor a 1.
  """

  if longitud < 1:
    raise ValueError("La longitud de la contraseña debe ser mayor o igual a 1.")

  caracteres = string.ascii_letters + string.digits + string.punctuation
  contraseña = "".join(random.choice(caracteres) for i in range(longitud))

  return contraseña


# Ejemplo de uso

longitud = int(input("Ingrese la longitud de la contraseña: "))
contraseña = generar_contraseña(longitud)

print(f"La contraseña generada es: {contraseña}")

```

```python
def calcular_imc():
    peso=float(input("Ingrese su peso en kilogramos: "))
    altura=float(input("Ingrese su altura en metros: "))
    imc=peso/(altura**2)
    imprime("Su IMC es: ", imc)
    si imc<18.5:
        imprime("Usted está bajo peso.")
    demás si 18.5<=imc<25:
        imprime("Usted tiene un peso normal.")
    demás si 25<=imc<30:
        imprime("Usted tiene sobrepeso.")
    demás:
        imprime("Usted tiene obesidad.")

calcular_imc()
```

```python
# Modifique en esta casilla el código anterior para que se ajuste al PEP8
"""
Módulo para calcular el índice de masa corporal (IMC).

Este módulo contiene una función para calcular el IMC de una persona a partir de
su peso en kilogramos y su altura en metros.

Funciones:
  calcular_imc(): Calcula el IMC de una persona.
"""

def calcular_imc():
  """
  Calcula el IMC de una persona a partir de su peso en kilogramos y su altura
  en metros.

  Argumentos:
    No hay.

  Retorno:
    No hay. Imprime el IMC y la categoría de peso de la persona.

  Excepciones:
    ValueError: Si el peso o la altura no son números válidos.
  """

  # Se solicita al usuario el peso
  try:
    peso = float(input("Ingrese su peso en kilogramos: "))
  except ValueError:
    print("Error: El peso ingresado no es un número válido.")
    return

  # Se solicita al usuario la altura
  try:
    altura = float(input("Ingrese su altura en metros: "))
  except ValueError:
    print("Error: La altura ingresada no es un número válido.")
    return

  # Se calcula el IMC
  imc = peso / (altura ** 2)

  # Se imprime el IMC
  print("Su IMC es: ", imc)

  # Se determina la categoría de peso
  if imc < 18.5:
    print("Usted está bajo peso.")
  elif 18.5 <= imc < 25:
    print("Usted tiene un peso normal.")
  elif 25 <= imc < 30:
    print("Usted tiene sobrepeso.")
  else:
    print("Usted tiene obesidad.")


# Se ejecuta la función para calcular el IMC
calcular_imc()
```

```python
# Modifique en esta casilla el código anterior para que se ajuste al PEP8
"""
Módulo para generar un patrón de asteriscos.

Este módulo contiene una función para generar un patrón de asteriscos
con un número determinado de filas.

Funciones:
  generar_patron(): Genera un patrón de asteriscos.
"""

def generar_patron(filas):
  """
  Genera un patrón de asteriscos con un número determinado de filas.

  Argumentos:
    filas: El número de filas del patrón.

  Retorno:
    No hay. Imprime el patrón de asteriscos.

  Excepciones:
    ValueError: Si el número de filas no es un número entero positivo.
  """

  # Se valida el número de filas
  if not isinstance(filas, int) or filas <= 0:
    raise ValueError("El número de filas debe ser un número entero positivo.")

  # Se genera el patrón
  for i in range(1, filas + 1):
    print("*" * i)


# Se solicita al usuario el número de filas
try:
  filas = int(input("Ingrese el número de filas del patrón: "))
except ValueError:
  print("Error: El valor ingresado no es un número válido.")
else:
  # Se genera el patrón
  generar_patron(filas)
```
